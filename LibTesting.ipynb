```python
from __future__ import print_function, division
%matplotlib inline
import thinkdsp
import thinkplot
import numpy as np
from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
from IPython.display import display
from IPython.display import Audio
```

```python
#Create the signals
sin1 = thinkdsp.SinSignal(freq=440, amp=1.0, offset=0)
sin2 = thinkdsp.SinSignal(freq=880, amp=0.5, offset=0)
sin3 = thinkdsp.SinSignal(freq=660, amp=0.3, offset=0)
sin4 = thinkdsp.SinSignal(freq=1360, amp=5, offset=0)
mix = sin1 + sin2 + sin3 + sin4

wave = mix.make_wave(duration=1, start=0, framerate=11025)
print(wave.ys)
audio = Audio(data=wave.ys, rate=wave.framerate)

#Take a N period of wave
#period = mix.period
#segment = wave.segment(start=0, duration=period*100)
#segment.normalize()
#segment.apodize()
#segment.plot()

#Cut of some frequencies
spectrum = wave.make_spectrum()
spectrum.low_pass(500) #Only low passes!
spectrum.plot(high=2000)

#Build the wave back!
filtered = spectrum.make_wave()
#filtered.normalize()
#filtered.apodize()

audio = Audio(data=wave.ys, rate=wave.framerate)
audio
```
